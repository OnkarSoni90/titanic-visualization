```python
# ðŸ“¦ Cell 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
# ðŸ“¦ Cell 2: Load Dataset
df = pd.read_csv("titanic.csv")
df.head()
```

```python
# ðŸ“¦ Cell 3: Overview
df.info()
```

```python
# ðŸ“¦ Cell 4: Descriptive Stats
df.describe(include='all')
```

```python
# ðŸ“Š Cell 5: Countplot â€“ Gender vs Survival
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival Count by Gender")
plt.show()
```

```python
# ðŸ“Š Cell 6: Barplot â€“ Survival by Class
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title("Survival Rate by Passenger Class")
plt.show()
```

```python
# ðŸ“Š Cell 7: Histogram â€“ Age Distribution
sns.histplot(df['Age'].dropna(), kde=True, bins=30)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# ðŸ“Š Cell 8: Boxplot â€“ Age vs Pclass
sns.boxplot(x='Pclass', y='Age', data=df)
plt.title("Age distribution by Passenger Class")
plt.show()
```

```python
# ðŸ“Š Cell 9: Heatmap â€“ Correlation
plt.figure(figsize=(10, 6))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# ðŸ“Š Cell 10: Pie Chart â€“ Gender Distribution
gender_counts = df['Sex'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Gender Distribution")
plt.show()
```
